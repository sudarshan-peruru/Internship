## Mission 8: Functions to make predictions

### Instructions

    â— Write a function named predict_price that can use the k-nearest neighbors machine
    learning technique to calculate the suggested price for any value for accommodates.
    This function should:
    
        â—‹ Take in a single parameter, new_listing, that describes the number of
        bedrooms.

        â—‹ We've added code that assigns dc_listings to a new Dataframe named temp_df.
        We used the pandas.DataFrame.copy() method so the underlying dataframe is
        assigned to temp_df, instead of just a reference to dc_listings.

        â—‹ Calculate the distance between each value in the accommodates column and the
        new_listing value that was passed in. Assign the resulting Series object to the
        distance column in temp_df.

        â—‹ Sort temp_df by the distance column and select the first 5 values in the
        pricecolumn. Don't randomize the ordering of temp_df.

        â—‹ Calculate the mean of these 5 values and use that as the return value for the
        entire predict_price function.
    
    â— Use the predict_price function to suggest a price for a living space that:
    
        â—‹ accommodates 1 person, assign the suggested price to acc_one.

        â—‹ accommodates 2 people, assign the suggested price to acc_two.

        â—‹ accommodates 4 people, assign the suggested price to acc_four.

```python
def predict_price(new_listing):
    temp_df = dc_listings
    temp_df['distance'] = np.sqrt((temp_df['accommodates'] - new_listing)**2)
    temp_df = temp_df.sort_values(by = ['distance'],ascending=True)
    result = temp_df['price'].head(5).mean()
    return result

acc_one = predict_price(1)
acc_two = predict_price(2)
acc_four = predict_price(4)
print(acc_one,acc_two,acc_four)
```
