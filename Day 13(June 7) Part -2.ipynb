## Mission 1: Testing Quality of Predictions

### Instructions

        ● Within the predict_price function, change the Dataframe that temp_df is assigned to.
        Change it from dc_listings to train_df, so only the training set is used.
        
        ● Use the Series method apply to pass all of the values in the accommodates column from
        test_df through the predict_price function.
        
        ● Assign the resulting Series object to the predicted_price column in test_df.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
def read(filename):
    df = pd.read_csv(filename)
    return df
df = read('Data Files/dc_listings.csv')
rand = np.random.permutation(df.index)
df = df.loc[rand]
df_train, df_test = train_test_split(df, test_size= 0.25, random_state = 0)
df_test.insert(2,'predicted_price',0,True)
# print(df_test.shape)
# print(df_train.shape)
# print(df.shape)
df_test
```

```python
pd.options.mode.chained_assignment = None
def predict_price(new_listing):
    temp_df = df_train
    temp_df['distance'] = np.sqrt((temp_df['accommodates'] - new_listing)**2)
    temp_df = temp_df.sort_values(by = ['distance'],ascending=True)
    result = temp_df['price'].head(5).mean()
    return result

df_test['predicted_price'] = df_test['accommodates'].apply(predict_price)
df_test[['accommodates','price','predicted_price']]
```

```python
mae = np.absolute((df_test['predicted_price'] - df_test['price']))
mae
```

## Mission 3: Mean Squared Error
### Instructions

        ● Calculate the MSE value between the predicted_price and price columns and assign to
        mse.

```python
pd.options.mode.chained_assignment = None
df_test['squared_error'] = (df_test['predicted_price'] - df_test['price'])**(2)
mse = df_test['squared_error'].mean()
rmse = mse ** (1/2)
mse
```
